```python
all_data = []
last_valid_profile = None  # Initialize the last valid profile
```

```python
start = 1 if last_valid_profile==None else last_valid_profile

for i in range(start, 3):
    try:
        data = get_pdb_data(profile=i)
        print(i)
        if data is not None:
            name = data['mbti_profile']
            mbti_df = data['breakdown_systems']['1']
            enneagram_df = data['breakdown_systems']['2']
            socionics_df = data['breakdown_systems']['3']
            alignement_df = data['breakdown_systems']['4']
            instinctual_variant_df = data['breakdown_systems']['5']
            tritype_df = data['breakdown_systems']['6']
            temperaments_df = data['breakdown_systems']['7']
            attitudnal_psyque_df = data['breakdown_systems']['8']
            big5_df = data['breakdown_systems']['9']

            t = [name, mbti_df, enneagram_df, socionics_df, alignement_df,
                 instinctual_variant_df, tritype_df, temperaments_df,
                 attitudnal_psyque_df, big5_df]

            all_data.append(t)
            last_valid_profile = i  # Update the last valid profile
        else:
            print(f"Profile {i} data is None. Skipping...")
    except Exception as e:
        print(f"Error fetching data for profile {i}: {e}")
        time.sleep(210.5)  # Wait for before the next request


# Now you can use 'all_data' for further processing
print(f"Last valid profile processed: {last_valid_profile}")
```
